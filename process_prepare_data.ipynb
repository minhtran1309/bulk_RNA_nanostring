```python
import pandas as pd
```

```python
xls = pd.ExcelFile('Data/LBL-10043-06_nCounter_PanCancer_Immune_Profiling_Panel_Gene_List.xlsx')
# df1 = pd.read_excel(xls, 'Sheet1')
# df2 = pd.read_excel(xls, 'Sheet2')
print(xls.sheet_names)
```

```python
gene_prob_details = pd.read_excel(xls, 'Gene and Probe Details', skiprows=[0, 773])
# df2 = pd.read_excel(xls, 'Sheet2')
gene_prob_details
```

```python
gannot = pd.read_excel(xls, 'Annotations', skiprows=[0,772])
# df2 = pd.read_excel(xls, 'Sheet2')
gannot
```

```python
values_not_in_df2 = gannot[~gannot['HUGO Name'].isin(gene_prob_details['Official Symbol'])]

print(values_not_in_df2)
```

```python
join_gene_annot = pd.merge(gannot, gene_prob_details, left_on='HUGO Name', right_on='Official Symbol')
join_gene_annot
```

```python
join_gene_annot.columns
```

```python
rename_col = {'HUGO Name':'Name', 'Gene Class':'Gene_Class'}
```

```python
join_gene_annot.rename(columns=rename_col, inplace=True)
```

```python
join_gene_annot.head()
```

```python
join_gene_annot.to_csv('Data/ProbeAnnot.csv')
```

```python
count_matrix = pd.read_csv('Data/230710_Tildra_Nanostring_Normalised_Data.csv')
```

```python
count_matrix
```

```python
count_matrix.columns
```

```python
gene_stat = count_matrix[['Probe Name', 'Class Name', 'Avg Count',
                           'Min Count','Max Count', '%CV']].copy()
gene_stat= gene_stat.drop([0,1])
gene_stat
```

```python
join_gene_annot.set_index('Name', inplace=True)
gene_stat.set_index('Probe Name', inplace=True)
```

```python
join_gene_annot.head()
```

```python
gene_stat.head()
```

```python
join_gene_stats = gene_stat.join(join_gene_annot,lsuffix="_left", rsuffix="_right", how='outer')
join_gene_stats
```

```python
join_gene_stats[join_gene_stats['Class Name'].isna()].index.values
```

```python
join_gene_stats_cleaned= join_gene_stats.drop(join_gene_stats[join_gene_stats['Class Name'].isna()].index.values)
join_gene_stats_cleaned.head()
```

```python
join_gene_stats_cleaned.columns
```

```python
gannot = join_gene_stats_cleaned[['Class Name', 'Avg Count', 'Min Count', 'Max Count', '%CV',
                         'Accession','Common Name','Gene_Class']]
gannot.to_csv('Data/gannot.csv')
```
