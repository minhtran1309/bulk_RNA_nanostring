```python

def sub_patient_id(patient_id):
    pattern = r"K\d{4}|K\d{3}"
#     print(patient_id)
    match = re.search(pattern, patient_id)
    
    if match:
#         print(match.group())
        return match.group()
    else:
#         print(patient_id)
        return None
```
